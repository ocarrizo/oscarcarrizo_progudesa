```python
def invertir_lista(lista):
  lista.reverse()
  return lista
```

```python
def collatz(nro):
  cont = 1
  while nro!= 1:
    if nro %2 ==0:
      nro= nro // 2
    else:
      nro= nro*3 + 1
      cont +=1
  return cont
```

```python
def contar_definiciones(d):
  for clave in d:
    d[clave] = len(d[clave])
  return d
```

```python
def cantidad_de_claves_letra(dd,l):
  cont = 0
  for clave in dd:
    if clave[0] == l:
      cont +=1
  return cont
```
